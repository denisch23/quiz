```python
import ipywidgets as widgets
from IPython.display import display, clear_output
```

```python
# Define questions
questions = [
    {"question": "What is the capital of France?", "options": ["Paris", "London", "Rome", "Berlin"], "answer": "Paris"},
    {"question": "What is 2 + 2?", "options": ["3", "4", "5", "6"], "answer": "4"}
]
```

```python
# Create a function to display a question
def display_question(q):
    print(q["question"])
    buttons = [widgets.Button(description=option) for option in q["options"]]

    def on_button_click(b):
        clear_output()
        if b.description == q["answer"]:
            print("Correct!")
        else:
            print("Wrong!")
        display_question(questions.pop(0)) if questions else print("Quiz completed!")

    for button in buttons:
        button.on_click(on_button_click)
        display(button)

display_question(questions.pop(0))
```
